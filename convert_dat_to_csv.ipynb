## This is to convert the .dat file to .csv file

```python
import os
import pandas as pd

def get_column_names():
    columns = [
        'timestamp', 'activity_id', 'heart_rate',
    ]
    imu_parts = ['hand', 'chest', 'ankle']
    imu_features = [
        'temp',
        'acc_x_16g', 'acc_y_16g', 'acc_z_16g',
        'acc_x_6g', 'acc_y_6g', 'acc_z_6g',
        'gyro_x', 'gyro_y', 'gyro_z',
        'mag_x', 'mag_y', 'mag_z',
        'orient_1', 'orient_2', 'orient_3', 'orient_4'
    ]
    for part in imu_parts:
        for feat in imu_features:
            columns.append(f'{part}_{feat}')
    return columns

def convert_dat_to_csv(data_folder):
    output_folder = 'csvData'
    os.makedirs(output_folder, exist_ok=True)
    columns = get_column_names()
    for filename in os.listdir(data_folder):
        if filename.endswith('.dat'):
            dat_path = os.path.join(data_folder, filename)
            csv_filename = os.path.splitext(filename)[0] + '.csv'
            csv_path = os.path.join(output_folder, csv_filename)
            df = pd.read_csv(dat_path, sep='\s+', header=None, names=columns)
            df.to_csv(csv_path, index=False)
            print(f'Converted {filename} to {csv_filename} in {output_folder}')

if __name__ == '__main__':
    data_folder = 'data'
    convert_dat_to_csv(data_folder)
```
